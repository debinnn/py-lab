(3) Write a program to create functions that can accept multiple dictionaries as arguments using ‘*args’, and perform various operations on them.

(a) Write a function, say, ‘merging_Dict(*args)’ that takes multiple dictionaries and merge them.\
(b) Write a function which can find common keys in multiple dictionaries.\
(c) Create a function that inverts a dictionary by swapping its keys and values. If multiple keys have the same value in the original dictionary, they should be grouped in a list as the value in the inverted dictionary.

Implement this program and demonstrate it by using adequate examples.

```python
def merging_Dict(*args):
    merged = {}
    for dictionary in args:
        merged.update(dictionary)
    return merged

def find_common_keys(*args):
    if not args:
        return set()
    common_keys = set(args[0].keys())
    for dictionary in args[1:]:
        common_keys.intersection_update(dictionary.keys())
    return common_keys

def invert_dictionary(dictionary):
    inverted = {}
    for key, value in dictionary.items():
        if value in inverted:
            if isinstance(inverted[value], list):
                inverted[value].append(key)
            else:
                inverted[value] = [inverted[value], key]
        else:
            inverted[value] = key
    return inverted

# Demonstration
if __name__ == "__main__":
    dict1 = {'a': 1, 'b': 2, 'c': 3}
    dict2 = {'b': 20, 'd': 4, 'e': 5}
    dict3 = {'c': 30, 'e': 5, 'f': 6}

    print("1. Merging dictionaries:")
    merged = merging_Dict(dict1, dict2, dict3)
    print("Merged dictionary:", merged)

    print("\n2. Finding common keys:")
    common = find_common_keys(dict1, dict2, dict3)
    print("Common keys:", common)

    print("\n3. Inverting a dictionary:")
    dict4 = {'a': 1, 'b': 2, 'c': 2, 'd': 3, 'e': 1}
    print("Original dictionary:", dict4)
    inverted = invert_dictionary(dict4)
    print("Inverted dictionary:", inverted)
```

(5) Write a Python program for a wether analyzing system to process and analyze weather data for Bangalore City from 20th July to 26th July in 2024.

(a) Each day’s data includes:
- Date
- Maximum temperature (in Celsius)
- Minimum temperature (in Celsius)
- Humidity (in percentage) 

(b) Write a function that finds the average maximum and minimum temperatures from the weather data.\
(c) Write a function that calculates the average humidity over the given period 

```python
weather_data = [
    {"date": "2024-07-20", "max_temp": 29, "min_temp": 21, "humidity": 65},
    {"date": "2024-07-21", "max_temp": 30, "min_temp": 22, "humidity": 70},
    {"date": "2024-07-22", "max_temp": 28, "min_temp": 20, "humidity": 75},
    {"date": "2024-07-23", "max_temp": 31, "min_temp": 23, "humidity": 60},
    {"date": "2024-07-24", "max_temp": 30, "min_temp": 22, "humidity": 68},
    {"date": "2024-07-25", "max_temp": 29, "min_temp": 21, "humidity": 72},
    {"date": "2024-07-26", "max_temp": 32, "min_temp": 24, "humidity": 55}
]

def average_temperatures(data):
    total_max = sum(day["max_temp"] for day in data)
    total_min = sum(day["min_temp"] for day in data)
    avg_max = total_max / len(data)
    avg_min = total_min / len(data)
    return avg_max, avg_min

def average_humidity(data):
    total_humidity = sum(day["humidity"] for day in data)
    return total_humidity / len(data)

def main():
    print("Weather Analysis for Bangalore City (July 20-26, 2024)")
    print("------------------------------------------------------")

    for day in weather_data:
        print(f"Date: {day['date']}, Max Temp: {day['max_temp']}°C, Min Temp: {day['min_temp']}°C, Humidity: {day['humidity']}%")

    avg_max, avg_min = average_temperatures(weather_data)
    print(f"\nAverage Maximum Temperature: {avg_max:.2f}°C")
    print(f"Average Minimum Temperature: {avg_min:.2f}°C")

    avg_humidity = average_humidity(weather_data)
    print(f"Average Humidity: {avg_humidity:.2f}%")

if __name__ == "__main__":
    main()
```
